```python
uneven = make_array(300, 350, 400, 500, 1800)
millions.bin('Adjusted Gross', bins=uneven)
```

```python
millions.hist('Adjusted Gross', bins=uneven, unit="Million Dollars")
```

```python
millions.hist('Adjusted Gross', bins=uneven, normed=False)
```

```python
histogram_elements = millions.bin('Adjusted Gross', bins=uneven).relabeled(1, 'count')
histogram_elements
```

```python
total_count = sum(histogram_elements.column('count'))
percents = histogram_elements.column('count')/total_count*100
histogram_elements = histogram_elements.with_columns('percent', percents)
histogram_elements
```

```python
bin_widths = np.diff(histogram_elements.column('bin'))
num_bins = histogram_elements.num_rows - 1    # the number of bins
histogram_elements = histogram_elements.take(
    np.arange(num_bins)).with_columns(
    'width', bin_widths
)
histogram_elements
```

```python
heights = np.round(
    histogram_elements.column('percent')/histogram_elements.column('width'),2)
histogram_elements = histogram_elements.with_columns('height', heights)
histogram_elements
```

```python
pokemon = Table.read_table('data/pokemon.csv')
pokemon
```

```python
# don't worry about this code
gen_types = pokemon.where('generation', are.below_or_equal_to(4)).pivot('generation', 'type1').relabeled(list(map(str, range(1, 5))), ['gen{} count'.format(i) for i in range(1, 5)])
gen_types
```

```python
gen_types.barh('type1')
```

```python
gen_types.select('type1', 'gen1 count', 'gen2 count').barh('type1')
```

```python
gen1_sum = sum(gen_types.column('gen1 count'))
gen2_sum = sum(gen_types.column('gen2 count'))
gen1_sum, gen2_sum
```

```python
gen_types.select('type1').with_columns(
    'gen1 percentage', gen_types.column('gen1 count') / gen1_sum * 100,
    'gen2 percentage', gen_types.column('gen2 count') / gen2_sum * 100
).barh('type1')
```
